```python
iris = datasets.load_iris()
```

```python
df = iris.data
df
```

```python
df = pd.DataFrame(df)
df
```

```python
#carregando classes
df1 = iris.target
df1
```

```python
#montando Dataframe
df1 = pd.DataFrame(df1)
df1
```

```python
#concatenando
df_iris = pd.concat([df,df1], axis=1)
df_iris
```

```python
#Carregando o nomes das colunas
colnames = iris.feature_names
colnames
```

```python
type(colnames)
```

```python
#Adicionando a classificalção da plantas
colnames.append('classification')
colnames
```

```python
#Atribuindo as colunas os nomes da colunas
df_iris.columns = colnames
df_iris
```

```python
#buscando os nomes das colunas
nomes = iris.target_names
nomes
```

```python
#Ordenando na coluna de classicação ordenando onde 0 setosa, 1 versicolor e 2 para virginica
df_iris['classification'] = df_iris['classification'].replace({0:'setosa',
                                                               1:'versicolor',
                                                               2:'virginica'})
    
df_iris
```
